```python
import matplotlib.pyplot as plt
import numpy as  np
import pandas as pd
import plotly.express as px
```

### 1. Line Plot (Trends & Time-Series)
Use Case: Stock market trends, website traffic, weather changes

Matplotlib 

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(12,4))
plt.plot(x, y, color='green', label='Sine Wave')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Matplotlib Line Plot")
plt.legend()
plt.show()
```

Plotly

```python
df = pd.DataFrame({'X': x, 'Y': y})
fig = px.line(df, 'X', 'Y', title= 'Plotly Line Plot', labels={'X':'X axis', 'Y': 'Y axis'})
fig.show()
```

### Bar Chart (Comparison & Categories)
Use Case: Sales reports, market share, survey results

Matplotlib

```python
types = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.figure(figsize=(12,5))
plt.bar(types, values, color=['red', 'blue', 'green', 'grey'])
plt.xlabel("Types")
plt.ylabel("Values")
plt.title("Matplotlib Bar Chart")
plt.show()
```

Plotly Bar Chart

```python
df = pd.DataFrame({'Types': types, 'Values':values})
fig = px.bar(df, 'Types', 'Values', title="Plotly Bar Chart", color='Types')
fig.show()
```

### Histogram (Distribution & Patterns)
Use Case: Income distribution, exam scores, error detection

Matplotlib

```python
data = np.random.randn(1000)

plt.figure(figsize=(12,6))
plt.hist(data, bins=30, color='blue', edgecolor="black")
plt.title("Matplotlib Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Plotly 

```python
df = pd.DataFrame({'Data': data})
fig = px.histogram(df,'Data',nbins=30, title='Plotly Histogram')
fig.show()
```

### Scatter Plot (Relationships & Correlations)
Use Case: Customer spending vs. income, marketing data

Matplotlib

```python
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(12,6))
plt.scatter(x, y, color="green")
plt.title("Matplotlib Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Plotly

```python
df = pd.DataFrame({'X': x, 'Y': y})
fig = px.scatter(df, 'X', 'Y', title="Plotly Scatter Plot", labels={'X': "X-axis", 'Y': "Y-axis"})
fig.show()
```

### Heatmap (Correlation & Relations)
Use Case: Weather maps, correlation matrices in finance

Plotly Heatmap

```python
np.random.seed(42)
data = np.random.rand(5,5)
df = pd.DataFrame(data, columns=["A", "B","C","D","E"])

fig = px.imshow(df, text_auto=True, color_continuous_scale='thermal', title='Plotly Heatmap')
fig.show()
```
